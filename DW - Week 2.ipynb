# Vector Calculations in Python with and without NumPy

```python
# Using plain Python to compute squares and cubes of integers from 0 to n

n = 3
a = [i**2 for i in range(n)] # Squares of integers
b = [i**3 for i in range(n)] # Cubes of integers

print("Vector a (squares):", a)
print("Vector b (cubes):", b)
```

```python
import numpy as np

# Now using NumPy to calculate same thing
n = 3
nums = np.arange(n) # Create array of integers from 0 to n
a = nums**2
b = nums**3

print("Vector a (squares):", a)
print("Vector b (cubes):", b)
```

# Iris Dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Iris dataset
iris = sns.load_dataset('iris')

#Print description
print(iris.describe(), "\n")

# Plot sepal length (column 1) as x and sepal width (column 2) as y
plt.scatter(iris['sepal_length'], iris['sepal_width'], color='blue', alpha=0.6)
plt.title("Iris Dataset: Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.grid(True)
plt.show()
```

# Boston Dataset

```python
# ImportError: 
# `load_boston` has been removed from scikit-learn since version 1.2.
# Had to find other means to load the data set

from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt

# Load Boston dataset
boston = fetch_openml(name = 'boston', version = 1, as_frame = True)

# Access the data and target
data = boston.data
target = boston.target

# Print description
print(data.describe())
```

```python
# Extract column 3 and column 5

x = data['INDUS']
y = data['NOX']

# Plot the data with each point marked as '+'
plt.plot(x, y, 'r+', markersize=6)
plt.title("Boston Dataset: Non-Retail Business Proportion vs Nitric Oxide Concentration")
plt.xlabel("Non-Retail Business Proportion")
plt.ylabel("Nitric Oxide Concentration")
plt.grid(True)
plt.show()
```

# Manipulating Array Shapes

```python
# New multi-dimensional array
array = np.arange(12).reshape(3,4)

print("Original array:\n", array, "\n")

# Ravel
raveled = array.ravel()
print("Raveled array:\n", raveled, "\n")

# Flatten
flattened = array.flatten()
print("Flattened array:\n", flattened, "\n")

# Set shape with tuple
array.shape = (4,3)
print("Reshaped array (4x3):\n", array, "\n")

# Transpose (switching rows and columns)
transposed = array.transpose()
print("Transposed array (back to original):\n", transposed, "\n")

# Resize
array.resize(6,2)
print("Resized array (6x2):\n", array)
```

# Array properties

```python
# Let's create a new array
array = np.arange(20).reshape(4,5)

print(array, "\n")

# Number of dimensions
print("Number of dimensions:", array.ndim)

# Count of elements
print("Number of elements:", array.size)

# Count of bytes
print("Bytes per element:", array.itemsize)

# Full count of bytes
print("Total bytes:", array.nbytes)
```
